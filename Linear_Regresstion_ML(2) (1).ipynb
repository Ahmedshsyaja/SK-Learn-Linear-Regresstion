```python
mean_x = np.mean(X)
mean_y = np.mean(Y)
m=len(X)
numer = 0
denom = 0
for i in range(m):
   numer +=(X[i]-mean_x)*(Y[i]-mean_y)
   denom +=(X[i]-mean_x) ** 2
   b1 = numer / denom
   b0 = mean_y -(b1 * mean_x)

print (b1,b0)
```

```python
max_x = np.max(X) + 100
min_x = np.min(X) - 100

x= np.linspace(min_x,max_x,1000)
y = b0 + b1 *x

plt.plot(x,y,color='#58b970',label='Regresstion line')
plt.scatter(X,Y,c='#ef5423',label='Scatter plot')
```
